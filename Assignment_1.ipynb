1. Implement 10 combinations of Data operators ( +, -, *, /, //, %) between 2 simple arrays.







```python
# Addition
arr1 = [10, 20, 30, 40, 50]
arr2 = [2, 4, 6, 8, 10]
for i in range(len(arr1)):
  for j in range(len(arr2)):
    print(arr1[i] + arr2[j])
```

```python
# Subraction
for i in range(len(arr1)):
  for j in range(len(arr2)):
    print(arr1[i] - arr2[j])
```

```python
#Multiplication alternate elements of arr1 with alternate elements of arr2, producing all combinations.
for i in range(0,len(arr1),2):
  for j in range(0,len(arr2),2):
    print(arr1[i] * arr2[j])
```

```python
# Division alternate elements of arr1 with alternate elements of arr2 producing of all combinations.
for i in range(1,len(arr1),2):
  for j in range(1,len(arr2),2):
    print(arr1[i] / arr2[j])
```

```python
# Modulas alternate elements of arr1 with alternate elements of arr2, producing all combinations.
for i in range(0,len(arr1),2):
  for j in range(0,len(arr2),2):
    print(arr1[i] % arr2[j])
```

```python
#Floor Division alternate elements of arr1 with alternate elements of arr2 producing of all combinations.
for i in range(1,len(arr1),2):
  for j in range(1,len(arr2),2):
    print(arr1[i] // arr2[j])
```

```python
# Exponent
for i in range(0,len(arr1),3):
  for j in range(0,len(arr2),3):
    print(arr1[i] ** arr2[j])
```

```python
#Arithmetic Operator Precedence without parantheses
for i in range(len(arr1)):
  for j in range(len(arr2)):
    result = arr1[i] + arr2[i] * 2
  print(result)
```

```python
#Arithmetic Operator Precedence with parantheses
for i in range(len(arr1)):
  for j in range(len(arr2)):
    result = (arr1[i] + arr2[i]) * 2
  print(result)
```

```python
# Exponent with parantheses
for i in range(len(arr1)):
  for j in range(len(arr2)):
    result = arr1[i] ** arr2[i] +2
  print(result)
```

```python
# Example of Arithmetic operator precedence
for i in range(len(arr1)):
  for j in range(len(arr2)):
    value = arr1[i] * arr2[i] / arr2[i] // arr2[i] % arr2[i]
  print(value)
```

2. Implement 2 examples each of " For " loop and " While " loop

```python
# Example of for loop
rows = 5

for i in range(1, rows + 1):
              # Print leading spaces
    for j in range(rows - i):
        print(" ", end=" ")
              # Print stars
    for k in range(2 * i - 1):
        print("*", end=" ")
            # Move to next line
    print()
```

```python
fruits=['apple','banana','cherry']
for fruit in fruits:
  print("I like",fruit)
```

```python
# Example of while loop
total = 0

while True:
    num = int(input("Enter a number (0 to stop): "))
    if num == 0:
        break
    total += num
    print(f"Current total: {total}")

print(f"Final total: {total}")
```
